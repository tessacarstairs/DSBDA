```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
sns.distplot(dataset['fare'])
```

```python
sns.distplot(dataset['fare'], kde=False)
```

```python
sns.jointplot(data=dataset, x='age',y='fare')
```

```python
sns.rugplot(dataset['fare'])
```

```python
sns.barplot(data=dataset,x='sex', y='age')
```

```python
sns.countplot(x='sex', data=dataset)
```

```python
sns.boxplot(data=dataset, x='sex', y='age')
```

```python
sns.boxplot(data=dataset, x='sex', y='age', hue='survived')
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
sns.stripplot(x='sex', y='age', data=dataset)
```

```python
sns.swarmplot(data=dataset, x='sex', y='age')
```

```python
sns.violinplot(data=dataset, x='sex', y='age')
sns.swarmplot(data=dataset, x='sex', y='age', color='black')
```

```python
titanic_hist = dataset.fare.plot.hist(bins = 40, color = 'grey')
plt.xlabel('Fare(pounds, 1912 prices)')

plt.show(titanic_hist)
```
